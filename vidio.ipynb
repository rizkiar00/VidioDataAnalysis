```
import pandas as pd
df = pd.read_csv('part-00000-4803092d-32ab-4912-b2ba-d37dd93537a2-c000.csv',error_bad_lines=False, warn_bad_lines=False) 
df = df.dropna(axis=1, how='all')
#df = df.sample(frac=0.05, random_state=1)
#df['play_time'] = df['play_time'].str.split('T')
#df['end_time'] = df['end_time'].str.split('T')
df
```

```
#df['genre_name'] = df['genre_name'].values.split(',').lower().sort()
a =  df['genre_name'].str.lower().fillna(' ')
a = a.str.split(',',expand=False)
a = a.apply(sorted)
df['genre_name'] = a
df
```

```
import numpy as np
v = df['genre_name'].values
l = [len(x) for x in v.tolist()]
f, u = pd.factorize(np.concatenate(v))
n, m = len(v), u.size
i = np.arange(n).repeat(l)

dummies = pd.DataFrame(
    np.bincount(i * m + f, minlength=n * m).reshape(n, m),
    a.index, u
)
del df['genre_name']
dummies
```

```
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn import preprocessing
pd.options.mode.chained_assignment = None 
#dfsample = df.sample(frac=0.2, random_state=1)
dfimportant = df[df.columns.difference(['hash_content_id','hash_play_id','hash_visit_id','hash_watcher_id','hash_event_id',
                                         'play_time','end_time','referrer','hash_film_id','title','film_title','app_version','browser_version',
                                         'os_version','season_name'])]




#sc = StandardScaler()
# get numeric data
num_X = dfimportant.select_dtypes(include=['float64','int64'])
# update the cols with their normalized values

min_max_scaler = preprocessing.MinMaxScaler()
dfimportant[num_X.columns] = min_max_scaler.fit_transform(num_X)
#replace NaN normalized numeric with -2
dfencode = dfimportant.fillna(0)
#select object data
dfimportanceobj = dfencode.select_dtypes(include=['object'])
#replace NaN with -1
dfimportanceobj = dfimportanceobj.fillna(0)
#Apply label encoding
le = preprocessing.LabelEncoder()
dfimportanceobj = dfimportanceobj.applymap(str).apply(le.fit_transform)
dfencode.loc[:,dfimportanceobj.columns] = dfimportanceobj
#dfencode = dfencode.join(dummies)
dfencode
```

```
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=3,subplot_titles=('How long a User watch a video ("play_duration")'
                                                    ,'Whatever a User "is_login" or not'
                                                    ,'Whatever a User "completed" a video or no'))

#Top 15 most important to how long a User watch a video ('play_duration')
y = df.iloc[:,df.columns == 'play_duration'].sample(frac=0.05, random_state=1)
x = dfencode[dfencode.columns.difference(['completed','is_login','play_duration'])].sample(frac=0.05, random_state=1)
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)

fit = bestfeatures.fit(x,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
#concat two dataframes for better visualization 
featureScoresplay_duration = pd.concat([dfcolumns,dfscores],axis=1)
featureScoresplay_duration.columns = ['Feature','Score']  #naming the dataframe columns

min_max_scaler = preprocessing.MinMaxScaler()
featureScoresplay_duration['Score'] = min_max_scaler.fit_transform(featureScoresplay_duration[['Score']])
x11 = go.Bar(x=featureScoresplay_duration.nlargest(15,'Score')['Feature'], y=featureScoresplay_duration.nlargest(15,'Score')['Score'])
fig.add_trace(x11,row=1, col=1)



#Top 15 most important Feature to whatever a User 'is_login' or not
y = df.iloc[:,df.columns == 'is_login'].sample(frac=0.05, random_state=1)
x = dfencode[dfencode.columns.difference(['completed','is_login','play_duration'])].sample(frac=0.05, random_state=1)
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)

fit = bestfeatures.fit(x,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
#concat two dataframes for better visualization 
featureScoresis_login = pd.concat([dfcolumns,dfscores],axis=1)
featureScoresis_login.columns = ['Feature','Score']  #naming the dataframe columns

min_max_scaler = preprocessing.MinMaxScaler()
featureScoresis_login['Score'] = min_max_scaler.fit_transform(featureScoresis_login[['Score']])
x12 = go.Bar(x=featureScoresis_login.nlargest(15,'Score')['Feature'], y=featureScoresis_login.nlargest(15,'Score')['Score'])
fig.add_trace(x12,row=1, col=2)



y = dfencode.iloc[:,dfencode.columns == 'completed'].sample(frac=0.05, random_state=1)
x = dfencode[dfencode.columns.difference(['completed','is_login','play_duration'])].sample(frac=0.05, random_state=1)
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)

fit = bestfeatures.fit(x,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
#concat two dataframes for better visualization 
featureScorescompleted = pd.concat([dfcolumns,dfscores],axis=1)
featureScorescompleted.columns = ['Feature','Score']  #naming the dataframe columns

min_max_scaler = preprocessing.MinMaxScaler()
featureScorescompleted['Score'] = min_max_scaler.fit_transform(featureScorescompleted[['Score']])
x13 = go.Bar(x=featureScorescompleted.nlargest(15,'Score')['Feature'], y=featureScorescompleted.nlargest(15,'Score')['Score'])
fig.add_trace(x13,row=1, col=3)

fig.update_layout(showlegend=False,title_text="15 most important Feature ")
fig.show()
```

```
import plotly.express as px

featureScorestotal = pd.DataFrame({'Feature':featureScorescompleted['Feature'],
                                   'Score': featureScorescompleted['Score'] + 
                                            featureScoresis_login['Score'] + 
                                            featureScoresplay_duration['Score']})
fig = px.bar(featureScorestotal.nlargest(15,'Score'), x='Feature', y='Score',title="How important a Feature to Label 'completed', 'is_login', 'play_duration' ")
fig.show()
```

```
fig = make_subplots(rows=3, cols=1,subplot_titles=('category_name'
                                                    ,'browser_name'
                                                    ,'platform'
                                                    ,'stream_type'))

dfcategory = df['category_name'].value_counts().reset_index().rename(
           columns={'index': 'category_name', 'category_name': 'count'}).nlargest(7,'count')
x11 = go.Bar(x=dfcategory['category_name'], y=dfcategory['count'])
fig.add_trace(x11,row=1, col=1)

dfbrowser = df['browser_name'].value_counts().reset_index().rename(
           columns={'index': 'browser_name', 'browser_name': 'count'}).nlargest(7,'count')
x21 = go.Bar(x=dfbrowser['browser_name'], y=dfbrowser['count'])
fig.add_trace(x21,row=2, col=1)

dfplatform = df['platform'].value_counts().reset_index().rename(
           columns={'index': 'platform', 'platform': 'count'}).nlargest(7,'count')
x31 = go.Bar(x=dfplatform['platform'], y=dfplatform['count'])
fig.add_trace(x31,row=3, col=1)


fig.update_layout(showlegend=False,height = 1000,title_text="4 Most Important Feature ")
fig.show()
```

```
dummies = dummies.rename(columns={" ": "NaN"})
dfencodegenre = dfencode.join(dummies)
dfencodegenre
```

```
dfiscompleted = dfencodegenre.loc[dfencodegenre['completed'] == 1]
dfiscompleted = dfiscompleted[dfiscompleted.columns.difference(x.columns.to_list()+['completed','is_login','play_duration'])]
dfnotcompleted = dfencodegenre.loc[dfencodegenre['completed'] == 2]
dfnotcompleted = dfnotcompleted[dfnotcompleted.columns.difference(x.columns.to_list()+['completed','is_login','play_duration'])]

dfislogin = dfencodegenre.loc[dfencodegenre['is_login'] == True]
dfislogin = dfislogin[dfislogin.columns.difference(x.columns.to_list()+['completed','is_login','play_duration'])]
dfnotlogin = dfencodegenre.loc[dfencodegenre['is_login'] == False]
dfnotlogin = dfnotlogin[dfnotlogin.columns.difference(x.columns.to_list()+['completed','is_login','play_duration'])]

dictcompleted = dfiscompleted.sum().to_dict()
dfnotcompleted = dfnotcompleted.sum().to_dict()
dfislogin = dfislogin.sum().to_dict()
dfnotlogin = dfnotlogin.sum().to_dict()
clientlist = dict([(k, [dictcompleted[k], dfnotcompleted[k],dfislogin[k],dfnotlogin[k]]) for k in dictcompleted])
clientlist
dfsumgenre = pd.DataFrame.from_dict(clientlist)
dfsumgenre.index = ['IsCompleted','NotCompleted','IsLogin','NotLogin']
#dfsumgenre.insert(loc=0, column='data', value=['iscompleted','notcompleted','islogin','notlogin'])
del dfsumgenre['NaN']
dfsumgenre
```
